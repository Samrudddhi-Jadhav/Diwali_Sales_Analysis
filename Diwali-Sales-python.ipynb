```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # visualisation data
import seaborn as sns
```

```python
# df = Data Frame
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape')
#to avoid encoding error(i.e: location path ensuring) use 'unicode_escape'
```

```python
#shape = counts no. of columns and no. of rows and shows.
df.shape
```

```python
# head() = by default show top first 5 rows
df.head()
```

```python
#info() = gives details of column and rows with details like datatype null not null all the necessory description
df.info()
```

```python
# drop([...]) = drop unrelated / blank cloumns
#axis=1 means meanse that vertical column
#inplace = True means save the changes
df.drop(['Status','unnamed1'], axis = 1, inplace=True)
```

```python
#check for null values
#sum = show no of null values that is count of null 
pd.isnull(df).sum()
```

```python
# dropna = Delete null values
# df_test.dropna(inplace= True) is the same to df_test = df_test.dropna()
df.dropna(inplace = True)
```

```python
# astype('name_datatype') to Change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
# get to know / show all the columns
df.columns
```

```python
#rename column
df.rename(columns = {'Marital_Statues':'Shadi'})
```

```python
#df.describe = method to return description of the data in the Dataframe(i.e: count mean, std, etc) based on numeric columns
df.describe()
```

```python
# use describe() for specific columns
df[['Age', 'Orders', 'Amount']].describe()
```

```python
# rename columns names....make them lower case and relace with underscore
#df.rename(columns = {'order Id' : 'order_id', 'City':'city'})
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.head()
```

```python
#using seaborn with sns
sns.countplot(x = 'gender', data = df)
```

```python
# container created to show values on bars
ax = sns.countplot(x = 'gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#group by based on gender same like sql 
df.groupby(['gender'], as_index=False)['amount'].sum().sort_values(by='amount', ascending=False)
```

```python
sales_gen = df.groupby(['gender'], as_index=False)['amount'].sum().sort_values(by='amount',ascending=False)
ax = sns.barplot(x = 'gender', y = 'amount', data = sales_gen)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#from above graph we can see that most of the buyers are females and even the purchasing power of females are greater then men.
```
